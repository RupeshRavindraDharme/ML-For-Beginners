```python
# importing libraries needed for the assignment
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the pumplins dataset
pumpkins = pd.read_csv('../data/US-pumpkins.csv')
pumpkins.head()
```

```python
#Cleaning the data
# removing the nan values
pumpkins.dropna(how='all', axis=1, inplace=True)
print(pumpkins.head())
pumpkins.shape
```

```python
colms = ['Type', 'Sub Variety', 'Origin District', 'Unit of Sale', 'Unnamed: 25']
pumpkins.drop(columns= colms, inplace=True)
print(pumpkins.head())
pumpkins.shape
```

```python
pumpkins.isna().sum()
```

```python
pumpkins.drop([ 'Mostly Low', 'Mostly High', 'Item Size' ],axis='columns', inplace=True)
pumpkins.dropna(axis = 0, inplace=True)
pumpkins.shape
```

```python
pumpkins.isna().sum()
```

```python
#Converting date to month
pumpkins['Month'] = pd.DatetimeIndex(pumpkins['Date']).month
pumpkins.head()
```

```python
pumpkins.drop('Date', axis = 1, inplace=True)
pumpkins['Price'] = (pumpkins['Low Price']+pumpkins['High Price'])/2
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case = True, regex = True)]
pumpkins.head()
```

```python
pumpkins.shape
```

```python
pumpkins.loc[pumpkins['Package'].str.contains('1 1/9'), 'Price'] =  pumpkins['Price']/1.1
pumpkins.loc[pumpkins['Package'].str.contains('1/2'), 'Price'] =  pumpkins['Price']/0.5
pumpkins.head() 
```

```python
# preprocessing 
from sklearn.preprocessing import LabelEncoder
pumpkins[['City Name', 'Package', 'Variety', 'Origin', 'Color', 'Repack']] = pumpkins[['City Name', 'Package', 'Variety', 'Origin', 'Color', 'Repack']].apply(LabelEncoder().fit_transform)
pumpkins.head()
```

```python
#Model creation and prediction
g = sns.PairGrid(pumpkins)
g.map(sns.scatterplot)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
y = pumpkins['Color']
X = pumpkins.drop('Color', axis= 1)
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LogisticRegression()
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)
accuracy_score(ypred, ytest)
```
