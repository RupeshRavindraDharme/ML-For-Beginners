# Nigerian Music scraped from Spotify - an analysis

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/nigerian-songs.csv')
df.head()
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
top = df.artist_top_genre.value_counts()
plt.figure(figsize = (10, 7))
sns.barplot(x = top.index[:5], y = top.values[:5])
plt.xticks(rotation = 45)
plt.title("Top genre", color = 'green')
```

```python
df = df[df.artist_top_genre != 'Missing']
top = df.artist_top_genre.value_counts()
plt.figure(figsize=(10, 7))
sns.barplot(x = top.index, y = top.values)
plt.xticks(rotation = 45)
plt.title('Top genre', color = 'aqua')
```

```python
df = df[(df['artist_top_genre'] == 'afro dancehall') | (df['artist_top_genre'] == 'nigerian pop') | (df['artist_top_genre'] == 'afropop')]
df = df[df.popularity >  0]
plt.figure(figsize= (10, 7))
top = df.artist_top_genre.value_counts()
sns.barplot(x = top.index, y = top.values)
plt.xticks(rotation = 45)
plt.title('Top genre', color = 'aqua')
```

```python
corrmat = df.corr()
plt.subplots(figsize = (12, 9))
sns.heatmap(corrmat, vmax=0.8, square= True)
plt.show()
```

```python
sns.set_theme(style='ticks')
g = sns.jointplot(
    data= df,
    x = 'popularity',
    y = 'danceability',
    hue = 'artist_top_genre',
    kind = 'kde'
)
```

```python
sns.FacetGrid(data=df, hue = 'artist_top_genre', height = 5)\
    .map(sns.scatterplot, 'popularity', 'danceability')\
    .add_legend()
```
